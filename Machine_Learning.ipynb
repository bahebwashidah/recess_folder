```python
df= pd.read_csv("heart_disease.csv")
df.head()
```

```python
#people with heart diseases
#target column
df.target.value_counts()
```

```python
#percentage of male and female
count_female=len(df[df.sex==0])
count_male=len(df[df.sex==1])
#show output
print('percentage of Female patients: {:.2f}%'.format(count_female/(len(df.sex))*100))
print('percentage of male patients: {:.2f}%'.format(count_male/(len(df.sex))*100))
```

```python
df.groupby('target').mean()
```

```python
#barplot showing heart disease frequency of age
pd.crosstab(df.age,df.target).plot(kind='bar',figsize=(20,6))
plt.title('heart disease freuency for ages')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#Data visualization of heart disease frequency for for sex
pd.crosstab(df.sex,df.target).plot(kind='bar',figsize=(20,6),color=["#228B22","#6F8FAF"])
plt.title('heart disease freuency for sex')
plt.xlabel("sex(0='Female',1='Male')")
plt.legend(['Have Disease','DoesNot have a Disease'])
plt.xticks(rotation=0)
plt.ylabel("Frequency")
plt.show()
```
